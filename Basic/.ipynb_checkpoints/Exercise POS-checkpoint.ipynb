```python
import nltk
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

# Read New Story

```python
with open("news_story.txt") as f:
    text = f.readlines()
text
```

```python
text = ' '.join(text)
text
```

# Print all Part of Speech

```python
doc = nlp(text)

for token in doc:
    print(token, "|", token.pos_)
```

# Print all NOUN

```python
doc = nlp(text)

nouns = []

for token in doc:
    if token.pos_  in ["NOUN"]:
        nouns.append(token)

nouns
```

# Print all Numbers

```python
doc = nlp(text)

numbers=[]

for token in doc:
    if token.pos_ in ["NUM"]:
        numbers.append(token)
numbers
```

# Count all POS

```python
count = doc.count_by(spacy.attrs.POS)
count
```

```python
for k,v in count.items():
    print(doc.vocab[k].text, v)
```
