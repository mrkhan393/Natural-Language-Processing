```python
import spacy

nlp = spacy.load("en_core_web_lg")
```

# Read Data

```python
import pandas as pd
```

```python
df = pd.read_json("news_dataset.json")
print(df.shape)
df.head()
```

```python
df.category.value_counts()
```

# Preprocessing Data

```python
def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)
    return " ".join(filtered_tokens)
```

```python
df['preprocessed_text'] = df.text.apply(preprocess)
```

```python
df.head()
```

# Extracting Vector from Text

```python
df['vector'] = df.preprocessed_text.apply(lambda text: nlp(text).vector)
```

```python
df.head()
```

# Model Training

## Label Mapping

```python
df['label_num'] = df.category.map({
    'BUSINESS': 0,
    'SPORTS': 1,
    'CRIME': 2,
    'SCIENCE': 3
})
```

```python
df.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.vector.values,
    df.label_num,
    test_size=0.2,
    random_state=2022,
    stratify = df.label_num
)
```

```python
X_train
```

```python
X_test
```

## Convert to 2D Data

```python
import numpy as np

X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
X_train_2d
```

```python
X_test_2d
```

## DecisionTreeClasifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

clf = DecisionTreeClassifier()
clf.fit(X_train_2d, y_train)

y_pred = clf.predict(X_test_2d)
print(classification_report(y_test, y_pred))
```

## Scale Data to fit positive values and MultinomialNB

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
```

```python
scaler = MinMaxScaler()
scaled_train_data = scaler.fit_transform(X_train_2d)
scaled_test_data = scaler.transform(X_test_2d)
```

```python
clf = MultinomialNB()
clf.fit(scaled_train_data, y_train)

y_pred = clf.predict(scaled_test_data)
print(classification_report(y_test, y_pred))
```

## KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_2d, y_train)

y_pred = knn.predict(X_test_2d)
print(classification_report(y_test, y_pred))
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train_2d, y_train)

y_pred = rf.predict(X_test_2d)
print(classification_report(y_test, y_pred))
```

## GradientBoostingClassifier and Confusion Matrix Heatmap

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(X_train_2d, y_train)

y_pred = gb.predict(X_test_2d)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='e')
plt.xlabel("Prediction")
plt.ylabel("Truth")
```
